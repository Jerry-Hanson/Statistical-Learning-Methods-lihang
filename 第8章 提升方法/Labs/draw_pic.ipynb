```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(error):
    return 1/2*np.log((1-error)/error)
```

```python
%config InlineBackend.figure_format = 'svg'

x = np.linspace(0.01,0.99,2000)
plt.figure(figsize = (8,4))
plt.scatter(x,f(x),c = 'blue')
plt.ylabel("$vote weight$")
plt.xlabel("$Total Error$")
plt.savefig("amount_of_say",dpi = 250)
```

```python
def v(a):
    return np.exp(a)
x = np.linspace(0,3.5,2000)
plt.figure(figsize = (8,4))
plt.scatter(x,v(x),c = 'blue')
plt.xlabel("$vote weight$")
plt.savefig("new_sample_weight_error",dpi = 250)
```

```python
def v(a):
    return np.exp(-a)
x = np.linspace(0,3.5,2000)
plt.figure(figsize = (8,4))
plt.scatter(x,v(x),c = 'blue')
plt.xlabel("$vote weight$")
plt.savefig("new_sample_weight_correct",dpi = 250)
```
