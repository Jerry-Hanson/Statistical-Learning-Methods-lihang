```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['label'] = iris.target
```

```python
data = np.array(df.iloc[:100, [0, 1, -1]])
X, y = data[:,:-1], data[:,-1]
y = np.array([1 if i == 1 else -1 for i in y])
```

```python
import numpy as np
class Perceptron(object):
    def __init__(self, learning_rate = 0.01, max_iter = 200):
        self.lr = learning_rate
        self.max_iter = max_iter

    def sign(self,x,w,b):
        y = np.dot(x,w) + b
        return y
    def fit(self, data, label):
        '''
        input:data(ndarray):训练数据特征
              label(ndarray):训练数据标签
        output:w(ndarray):训练好的权重
               b(ndarry):训练好的偏置
        '''
        #编写感知机训练方法，w为权重，b为偏置
        self.w = np.array([1.]*data.shape[1])
        self.b = np.array([1.])
        #********* Begin *********#
        is_wrong = False
        while not is_wrong:
            wrong_count = 0
            for d in range(len(data)):
                X = data[d]
                y = label[d]
                if y * self.sign(X,self.w,self.b) <= 0:
                    self.w = self.w + self.lr * np.dot(y, X)
                    self.b = self.b + self.lr * y
                    wrong_count += 1
            if wrong_count == 0:
                is_wrong = True

        #********* End *********#
    def predict(self, data):
        '''
        input:data(ndarray):测试数据特征
        output:predict(ndarray):预测标签
        '''
       
        #********* Begin *********#
        predict = []
        for d in range(len(data)):
            X = data[d]
            if sign(X,self.w,self.b) <=0:
                predict.append(-1)
            else:
                predict.append(1)
        #********* End *********#
        return predict
    
```
